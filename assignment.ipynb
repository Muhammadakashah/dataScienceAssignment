```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
# Load the dataset
df = pd.read_csv("/content/Transparency_Code_Fraud_Data.csv",encoding='cp1252')
def fraud_Data():
  df=df.dropna()
  #defininng the variable for data presentation
  data = df['Effective Date']
  fec = df['FTE fraud employees count']
  fsc = df['FTE fraud specialist count']


# Plot the data
plt.plot(data,fec, label='FTE fraud employees count')
plt.plot(data,fsc,label='FTE fraud specialist count')
plt.plot(data,tfsc,label='Total fraud specialists count')

plt.legend()
plt.title("Fraud Data")
plt.xlabel('Years')
plt.ylabel('Values')
plt.grid(False)
plt.tight_layout()
plt.savefig('furad.png')
plt.show()
```
